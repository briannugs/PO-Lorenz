```julia
function DoubPen(t,x)
    #this function defines the motion of a double pendulum using the Lagrangian
    #x = [θ1 θ2 ω1 ω2]
    #I need to definel1,l2,m1,m2
    l1 = 2
    l2 = 1
    m1 = 3
    m2 = 1
    
    dx = zeros(4)
    
    dx[1] = x[3]
    dx[2] = x[4]
    
    
    dx[3] = (1/(l1*(1+m1-cos(x[1]-x[2])^2)))*(sin(x[1]-x[2])*(l1*cos(x[1]-x[2])*x[3]^2+x[4]^2)-(0.5*((1+2*m1)*sin(x[1])+sin(x[1]-2*x[2]))))
    
    dx[4] = sin(x[1]-x[2])*(((1+m1)*(cos(x[1]+l1*x[3]^2))+cos(x[1]-x[2])*x[4]^2))/(1+m1-cos(x[1]-x[2])^2)
    
    dx
end
```

```julia
function rungekutta4(f, y₀, Δt, t₀, t₁) 
   
    # fix up Δt so it evenly divides time interval
    N = convert(Int64, round((t₁ - t₀)/Δt))
    Δt = (t₁ - t₀)/N
    Δt2 = Δt/2
    Δt6 = Δt/6
    
    t = range(t₀, stop=t₁, length=N+1)

    y = zeros(N+1, length(y₀))
    y[1,:] = y₀

    for n = 1:N
        yn = y[n,:]
        tn = t[n]
        s1 = f(tn, yn)
        s2 = f(tn + Δt2, yn + Δt2*s1)
        s3 = f(tn + Δt2, yn + Δt2*s2)
        s4 = f(tn + Δt,  yn + Δt *s3)

        y[n+1, :] =  yn + Δt6*(s1+2s2 + 2s3 + s4) # RK4 update formula
    end
    t,y
end
```

```julia
function recurrence(x0,t,T)
    #This function will create a recurrence plot of the double pendulum system for some time, t, and some period, T. 
    
    #First the function will be evaluated at t and t+T
    Δt = 0.01
    t0 = 0
    tf = t+T
    tT,yT = rungekutta4(DoubPen,x0,Δt,t0,tf) 
    
    tt,yt = rungekutta4(DoubPen,x0,Δt,t0,t)
    
    #Now a norm of ||f(t)-f(t+T)|| will be calculated and put into a matrix R
    R = zeros(t,T)
    
    for i = 1:t
        for j = 1:T
            
            r = yt[i,:]-yT[i+j,:]
            
            R[i,j] = sqrt(r[1]^2+r[2]^2+r[3]^2+r[4]^2)
            
        end
    end 
    
    #The recurrence plot is created here along with a time series for some time,t
    P1 = contour(R',fill=true)
    P2 = plot(tt,yt[:,1])
    plot(P1,P2)
end
```
